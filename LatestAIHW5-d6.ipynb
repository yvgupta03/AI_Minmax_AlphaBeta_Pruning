```python
pos_dict = {}

for i in range(0,10):
    for j in range(0,6):
        pos_dict[(j,i)] = 0
```

```python
goal_pos = (5,9)
blue_pos = (5,0)
red_pos = (0,4)
```

```python
def neighbors(pos,temp_red, temp_blue):
    lst=[(pos[0],pos[1]+1),(pos[0],pos[1]-1),(pos[0]-1,pos[1]),(pos[0]+1,pos[1])]
    answer=[]
    for p in lst:
        if(p == temp_blue):
            continue
        elif(p == temp_red):
            continue
        elif( (p[0] < 0 or p[0] > 5 or p[1] < 0 or p[1] > 9) or (p[0] == 5 and (p[1] >= 1 or p[1] <= 8))):
            continue
        answer.append(p)
    
    return answer
```

```python
def min_max():
    
    global blue_pos
    global red_pos
    global goal_pos
    
    count = 0
    
    while((blue_pos != goal_pos) and (count < 100)):
        
        count += 1
        
        #print(count)
        
        answer = neighbors(blue_pos, red_pos, blue_pos)
        
        if len(answer) == 1:    #Selecting the only neighbor available and no need of creating a tree for that neighbor
            blue_pos = answer[0]
        elif len(answer) >= 2:
            temp_max = float('-inf')
            temp_max_pos = tuple()
            for p in answer:
                v = create_tree_blue(0, "min", p, red_pos)  ## we get min for all subttrees and then will calculate max 
                if v >= temp_max:
                    temp_max = v
                    temp_max_pos = p
            pos_dict[blue_pos] = 0
            blue_pos = temp_max_pos
            pos_dict[blue_pos] = 1
            
        if(blue_pos == goal_pos):
            print("Hurrrayyyyy")
            return
        
        answer = neighbors(red_pos, red_pos, blue_pos)
        
        if len(answer) == 1:    #Selecting the only neighbor available and no need of creating a tree for that neighbor
            red_pos = answer[0]
        elif len(answer) >= 2:
            temp_max = float('inf')
            temp_max_pos = tuple()
            for p in answer:
                v = create_tree_blue(0, "max", blue_pos, p)  ## we get min for all subttrees and then will calculate max 
                if v <= temp_max:
                    temp_max = v
                    temp_max_pos = p
            pos_dict[red_pos] = 0
            red_pos = temp_max_pos
            pos_dict[red_pos] = 2
        
        display(pos_dict)
        print("Next")
```

```python
def evaluate(blue_pos, red_pos):
    
    mdist = abs(goal_pos[0] - blue_pos[0]) + abs(goal_pos[1] - blue_pos[1])
    vdist = abs(blue_pos[0] - red_pos[0])
    
    return (-mdist)+vdist
```

```python
def create_tree_blue(d,level_type, parent_blue, parent_red):
    
    if(d == 6):
        ## fucntion to calculate utility function
        return evaluate(parent_blue, parent_red)
    
    if(d < 6):
        if(level_type == "min"):
            answer = neighbors(parent_red, parent_blue, parent_red)
            temp_v = float('inf')
            temp_pos = tuple()
            for p in answer:
                v = create_tree_blue(d+1,"max",parent_blue, p)
                if(v <= temp_v):
                    temp_v = v
                    temp_pos = p
            return temp_v
        elif(level_type == "max"):
            answer = neighbors(parent_blue, parent_blue, parent_red)
            temp_v = float('-inf')
            temp_pos = tuple()
            for p in answer:
                v = create_tree_blue(d+1,"min",p, parent_red)
                if(v >= temp_v):
                    temp_v = v
                    temp_pos = p
            return temp_v
                    
        
```
